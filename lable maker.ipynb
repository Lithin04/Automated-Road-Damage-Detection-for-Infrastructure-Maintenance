```python
import xml.etree.ElementTree as ET
import os
```

```python
def convert_annotation(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    labels = []
    for obj in root.findall('object'):
        class_name = obj.find('name').text
        if(class_name=='D00'):
            class_name='0'
            xmin = int(obj.find('bndbox/xmin').text)
            ymin = int(obj.find('bndbox/ymin').text)
            xmax = int(obj.find('bndbox/xmax').text)
            ymax = int(obj.find('bndbox/ymax').text)

        # Calculate bounding box center and dimensions
            width = xmax - xmin
            height = ymax - ymin
            x_center = (xmin + xmax) / 2.0
            y_center = (ymin + ymax) / 2.0

        # Normalize coordinates
            img_width = int(root.find('size/width').text)
            img_height = int(root.find('size/height').text)
            x_center /= img_width
            y_center /= img_height
            width /= img_width
            height /= img_height

        # Format label for YOLO: class_id x_center y_center width height
            label = f"{class_name} {x_center} {y_center} {width} {height}"
            labels.append(label)
        if(class_name=='D10'):
            class_name='1'
            xmin = int(obj.find('bndbox/xmin').text)
            ymin = int(obj.find('bndbox/ymin').text)
            xmax = int(obj.find('bndbox/xmax').text)
            ymax = int(obj.find('bndbox/ymax').text)

        # Calculate bounding box center and dimensions
            width = xmax - xmin
            height = ymax - ymin
            x_center = (xmin + xmax) / 2.0
            y_center = (ymin + ymax) / 2.0

        # Normalize coordinates
            img_width = int(root.find('size/width').text)
            img_height = int(root.find('size/height').text)
            x_center /= img_width
            y_center /= img_height
            width /= img_width
            height /= img_height

        # Format label for YOLO: class_id x_center y_center width height
            label = f"{class_name} {x_center} {y_center} {width} {height}"
            labels.append(label)
        if(class_name=='D20'):
            class_name='2'
            xmin = int(obj.find('bndbox/xmin').text)
            ymin = int(obj.find('bndbox/ymin').text)
            xmax = int(obj.find('bndbox/xmax').text)
            ymax = int(obj.find('bndbox/ymax').text)

        # Calculate bounding box center and dimensions
            width = xmax - xmin
            height = ymax - ymin
            x_center = (xmin + xmax) / 2.0
            y_center = (ymin + ymax) / 2.0

        # Normalize coordinates
            img_width = int(root.find('size/width').text)
            img_height = int(root.find('size/height').text)
            x_center /= img_width
            y_center /= img_height
            width /= img_width
            height /= img_height

        # Format label for YOLO: class_id x_center y_center width height
            label = f"{class_name} {x_center} {y_center} {width} {height}"
            labels.append(label)
        if(class_name=='D40'):
            class_name='3'
            xmin = int(obj.find('bndbox/xmin').text)
            ymin = int(obj.find('bndbox/ymin').text)
            xmax = int(obj.find('bndbox/xmax').text)
            ymax = int(obj.find('bndbox/ymax').text)

        # Calculate bounding box center and dimensions
            width = xmax - xmin
            height = ymax - ymin
            x_center = (xmin + xmax) / 2.0
            y_center = (ymin + ymax) / 2.0

        # Normalize coordinates
            img_width = int(root.find('size/width').text)
            img_height = int(root.find('size/height').text)
            x_center /= img_width
            y_center /= img_height
            width /= img_width
            height /= img_height

        # Format label for YOLO: class_id x_center y_center width height
            label = f"{class_name} {x_center} {y_center} {width} {height}"
            labels.append(label)

    return labels
```

```python
# Example usage:
xml_file = "D:/trinit/Czech/train/annotations/xmls/Czech_003579.xml"  # Path to your XML file
labels = convert_annotation(xml_file)
for label in labels:
    print(label)
```

```python
# Directory containing XML annotation files
xml_dir = "D:/trinit/Japan/train/annotations/xmls"

# Directory to save YOLO format text files
output_dir = "D:/trinit/Japan/train/labels"

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate through each XML file in the directory
for xml_file in os.listdir(xml_dir):
    if xml_file.endswith(".xml"):
        # Convert annotation to YOLO format
        labels = convert_annotation(os.path.join(xml_dir, xml_file))

        # Save YOLO format labels to a text file
        txt_file = os.path.splitext(xml_file)[0] + ".txt"
        with open(os.path.join(output_dir, txt_file), 'w') as f:
            for label in labels:
                f.write(label + '\n')
```
